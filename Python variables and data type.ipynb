## Python variables and data type

```python
# string
name = 'Alice'
company = 'XYZ'
print("String Example")
print(f"name :{name}")
print(f"company:{company}")
print(f"Type of Name: {type(name)}")
```

```python
# integer variable
age = 25
year = 2025
number = 23
print(age)
print(year)
print(number)
print(f"type of age:{type(age)}")
```

```python
#float variable
height = 5.7
weight = 50.8
temperature = 10.8
pi = 3.14
print("Float Examples")
print(height)
print(weight)
print(temperature)
print(pi)
print(f"Type of pi:{type(pi)}")
```

```python
#Boolean variables 
is_student = True
is_employeed = False
print(is_student)
print(is_employeed)
print(f"TYpe of student:{type(is_student)}")
```
